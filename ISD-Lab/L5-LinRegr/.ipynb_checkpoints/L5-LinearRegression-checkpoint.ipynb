### 1.5) Model using Simple Linear Regression

**NOTES :**
On our plots, RED line represents the sklearn Linear Regression and the GREEN represents our Linear Regression

```python
## Step by step : Gradient method
## Ref. pg 31 of 9.ISD_RegressionLineaire.pdf
## MSE : Mean Squared Error
def compute_MSE(b, m, data):
    totalError = 0
    for i in range(0, len(data)):
        x = data[i, 0]
        y = data[i, 1]
        totalError += (y-(m*x + b)) ** 2 
    return totalError / float(len(data))

def step_gradient(b_current, m_current, data, learning_rate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(data))
    for i in range(0, len(data)):
        x = data[i, 0]
        y = data[i, 1]
        b_gradient += -(1/N) * (y - (m_current * x + b_current))
        m_gradient += -(1/N) * x * (y - (m_current * x + b_current))
    
    new_b = b_current - (learning_rate * b_gradient)
    new_m = m_current - (learning_rate * m_gradient)
    return [new_b, new_m]

def gradient_descent(data, starting_b, starting_m, learning_rate, num_epochs):
    b = starting_b
    m = starting_m
    for i in range(num_epochs):
        b, m = step_gradient(b, m, np.array(data), learning_rate)
        #print("After {0} epochs : b = {1}, m = {2}, error = {3}".format(i, b, m, compute_MSE(b, m, data)))
    return [b, m]
```

```python
# Step 1 : Normalization (values between 0 & 1)
dfNorm = (df - df.min()) / (df.max() - df.min())
# Check normalization
#print(dfNorm)

# Step 2 : Hyper-parameters selection
learning_rate = 0.03
num_epochs = 500

# Step 3 : "m" and "b" init
initial_b = np.random.rand(1)
initial_m = np.random.rand(1)

# Step 4 : Application
print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_MSE(initial_b, initial_m, np.array(dfNorm[['RM','MEDV']]))))
print("Running ...")
[b, m] = gradient_descent(np.array(dfNorm[['RM','MEDV']]), initial_b, initial_m, learning_rate, num_epochs)
print("After {0} epochs : b = {1}, m = {2}, error = {3}".format(num_epochs, b, m, compute_MSE(b, m, np.array(dfNorm[['RM','MEDV']]))))
```

```python
# Regression from sklearn
linRegr = LinearRegression()

# Recovering values of a specific attribut (in this case : RM)
x_base = np.array(dfNorm['RM']).reshape(-1,1)
y_base = np.array(dfNorm['MEDV']).reshape(-1,1)

linRegr.fit(x_base, y_base)
y_predict_sklearn = linRegr.predict(x_base)

plt.scatter(x_base, y_base)
# Plot of the sklearn lib.
plt.plot(x_base, y_predict_sklearn, 'r')
# Plot using the return values b & m
plt.plot(x_base, m * x_base + b, 'xkcd:baby poop green')
plt.show()
```

```python
def sklearn_y_predict_recovery(attribut):
    linRegr = LinearRegression()
    x_base = np.array(dfNorm[attribut]).reshape(-1,1)
    y_base = np.array(dfNorm['MEDV']).reshape(-1,1)
    linRegr.fit(x_base, y_base)
    return linRegr.predict(x_base)
```

```python
def train_test_split(dataset, split=0.80): # training to 0.80 of dataset as asked in the lab sheet
    train = list()
    train_size = split * len(dataset)
    dataset_copy = list(dataset)    
    while len(train) < train_size:
        idx = random.randrange(len(dataset_copy))
        train.append(dataset_copy.pop(idx))
    return train, dataset_copy
```

```python
def simpleRegression(learning_rate, num_epochs, attribut, b_in = np.random.rand(1), m_in = np.random.rand(1)):
    # Take attribut values from normalized df
    data = dfNorm[[attribut,'MEDV']]
    
    # Step 3 : "m" and "b" init
    initial_b = np.random.rand(1)
    initial_m = np.random.rand(1)
    
    # Make our df into a numpy array for gradient fn()
    data = np.array(data)
    # Split our array into training and test set
    train, test = train_test_split(data)
    # Convert train and test into np.array()
    train = np.array(train)
    train.reshape(-1, 2)
    test = np.array(test)
    test.reshape(-1, 2)
    

    linRegr.fit(train[:,0].reshape(-1,1), train[:,1].reshape(-1,1))
    y_predicted = linRegr.predict(data[:,0].reshape(-1,1))
    y_origin = data[:,1]

    #print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_MSE(initial_b, initial_m, np.array(dfNorm[['RM','MEDV']]))))
    #print("Running ...")
    [b, m] = gradient_descent(np.array(dfNorm[[attribut,'MEDV']]), initial_b, initial_m, learning_rate, num_epochs)
    #print("After {0} epochs : b = {1}, m = {2}, error = {3}".format(num_epochs, b, m, compute_MSE(b, m, np.array(dfNorm[['RM','MEDV']]))))
    return [data, b, m, y_origin, y_predicted]
    
```

```python
# Test simpleRegression to be used in the final one (linearRegression())
simpleRegression(learning_rate, num_epochs, 'RM')

plt.scatter(x_base, y_base)
# Plot of the sklearn lib.
plt.plot(x_base, y_predict_sklearn, 'r')
# Plot using the return values b & m
plt.plot(x_base, m * x_base + b, 'xkcd:baby poop green')
plt.show()
```

```python
def linearRegression(learning_rate, num_epochs, attribut, nbr_of_split_part = 100):
    b_total = 0
    m_total = 0
    for i in range(nbr_of_split_part):
        [res, b_res, m_res, y_orgn, y_predict] = simpleRegression(learning_rate, num_epochs, attribut)
        b_total += b_res
        m_total += m_res
        
    b_avg = b_total / nbr_of_split_part # Average of b
    m_avg = m_total / nbr_of_split_part # Average of m
    
    plt.scatter(res[:,0], y_orgn)
    plt.title('Linear Regression on var. : {0} over MEDV'.format(attribut))
    plt.xlabel(attribut)
    plt.ylabel('MEDV')
    # Plot of the sklearn lib.
    plt.plot(res[:,0], y_predict_sklearn, 'r')
    # Plot using the return values b & m
    y_predict_from_gradient = m_avg * res[:,0] + b_avg
    plt.plot(res[:,0], y_predict_from_gradient, 'xkcd:baby poop green')
    plt.show()
    
    return y_predict_from_gradient
```

```python
learning_rate, num_epochs = 0.03, 500
working_attribut = 'RM'

y_predict_sklearn = sklearn_y_predict_recovery(working_attribut)

#y_pred = linearRegression(learning_rate, num_epochs, working_attribut, 1)
y_pred = linearRegression(learning_rate, num_epochs, working_attribut)

print('r2_score (sklearn) : {:.6f}'.format(sklearn.metrics.r2_score(dfNorm['MEDV'], y_predict_sklearn)))
print('r2_score (our)     : {:.6f}'.format(sklearn.metrics.r2_score(dfNorm['MEDV'], y_pred)))
```

### 2.2) Use sklearn lib. to prepare the boston df and then use the Hold Out validation to find the r2_score.

```python
# Reset x_values to only 'RM' & 'LSTAT' and y_value to 'MEDV'
x_values = np.array(dfNorm[['RM', 'LSTAT']])
y_value = np.array(dfNorm['MEDV'])

for tries in [100, 200, 300, 400, 500]:
    # Add-on values : 
    best_score = 0
    best_i = 0
    score_sum = 0
    score_avg = 0
    for i in range(tries):
        X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x_values, y_value, test_size=0.8, random_state=i)

        sk_learn_score = LinearRegression().fit(x_values, y_value).score(X_test, y_test)
        score_sum += sk_learn_score

        if sk_learn_score > best_score: 
            best_score = sk_learn_score 
            best_i = i

    score_avg = score_sum / tries

    print('With {} tries, the average score is : {:.6f}'.format(tries, score_avg))
    print('The best score is                    : {:.6f}'.format(best_score))
    print()
```

```python
# Reset x_values to only 'RM' & 'LSTAT' and y_value to 'MEDV'
x_values = np.array(dfNorm[['RM', 'LSTAT']])
y_value = np.array(dfNorm['MEDV'])

for tries in [100, 200, 300, 400, 500]:
    # Add-on values : 
    best_score = 0
    best_i = 0
    score_sum = 0
    score_avg = 0
    for i in range(tries):
        X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x_values, y_value, test_size=0.1, random_state=i)

        sk_learn_score = LinearRegression().fit(x_values, y_value).score(X_test, y_test)
        score_sum += sk_learn_score

        if sk_learn_score > best_score: 
            best_score = sk_learn_score 
            best_i = i

    score_avg = score_sum / tries

    print('With {} tries, the average score is : {:.6f}'.format(tries, score_avg))
    print('The best score is                    : {:.6f}'.format(best_score))
    print()
```

### 2.3) Results (based on pt 1 + 2)

We can see that a hundred tries is enough to evaluate the average score, but with more tries, we have more chances to find a higher best score.
Our average is around 0.62 which is better than when we used only one variable.

Plus, we test two differents values of splitting test (0.8 & 0.1) and we observed that with 0.8, we have a higher average score than with 0.1. But we have a higher best score with 0.1 than with 0.8.
